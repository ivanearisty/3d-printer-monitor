```python
from bambulabs_api import Printer
from os import path, environ
```

```python
from dotenv import load_dotenv
load_dotenv()

PRINTER_ACCESS_CODE = environ.get('PRINTER_ACCESS_CODE')
PRINTER_SERIAL_NUMBER = environ.get('PRINTER_SERIAL_NUMBER')
PRINTER_IP_ADDRESS = environ.get('PRINTER_IP_ADDRESS')

if not PRINTER_ACCESS_CODE:
    raise ValueError("PRINTER_ACCESS_CODE environment variable is required")
if not PRINTER_SERIAL_NUMBER:
    raise ValueError("PRINTER_SERIAL_NUMBER environment variable is required")
if not PRINTER_IP_ADDRESS:
    raise ValueError("PRINTER_IP_ADDRESS environment variable is required")

printer = Printer(access_code=PRINTER_ACCESS_CODE,
                  serial=PRINTER_SERIAL_NUMBER,
                  ip_address=PRINTER_IP_ADDRESS)
```

```python
printer.connect()
print(printer)
```
